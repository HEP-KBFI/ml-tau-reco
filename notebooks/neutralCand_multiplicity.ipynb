```python
import vector
import awkward as ak
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
data = []
data_qcd = []
for f in glob.glob('/local/snandan/HPS_correctpath/HPS/ZH_Htautau//*parquet')[:100]:
    data.append(ak.Array(ak.from_parquet(f).tolist()))
for f in glob.glob('/local/snandan/HPS_correctpath/HPS/QCD/*parquet')[:100]:
    data_qcd.append(ak.Array((ak.from_parquet(f).tolist())))
data = ak.concatenate(data)
data_qcd = ak.concatenate(data_qcd)
print('data loaded')
```

```python
zh_pt = data.tau_neutralHadron_pt
#print(ak.any(zh_pt==0))
a,b,c=plt.hist(
        ak.flatten(data.tau_neutralHadron_pt), histtype='step', label='ZH_tautau, neutral hadron pt with no cut', lw=2, bins=100,\
        range=(0,50),
        alpha=1, stacked=True, density=True, log=True
    )
a,b,c=plt.hist(
        ak.flatten(data_qcd.tau_neutralHadron_pt), histtype='step', label='QCD, neutral hadron pt with no cut', lw=2,\
        bins=100, range=(0,50),
       alpha=1, stacked=True, density=True, log=True
    )
plt.legend()
```
